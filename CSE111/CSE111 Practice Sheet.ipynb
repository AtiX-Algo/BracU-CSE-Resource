```python
class myList:
    def __init__(self, *args):
        self.values = list(args)

    def sum(self):
        total = 0
        for value in self.values:
            total += value
        print("Sum:", total)

    def merge(self, *args):
        self.values.extend(args)

    def average(self):
        if not self.values:
            print("Average: 0")
        else:
            avg = sum(self.values) / len(self.values)
            print("Average:", avg)


# Driver Code
l1 = myList(2, 3, 4, 5, 6)
l1.sum()
l1.merge(4, 5, 9)
l1.sum()
l1.average()
print("-----------------------------")
l2 = myList()
l2.average()
l2.merge(1, 2, 4, 8)
l2.sum()
```

```python
class Student:
    def __init__(self, name, ID, semester):
        self.name = name
        self.ID = ID
        self.semester = semester

    def Details(self):
        return "Name: " + self.name + "\n" + "ID: " + self.ID + "\nCurrent semester: " + self.semester + "\n"


class CSEStudent(Student):
    def __init__(self, name, ID, semester):
        super().__init__(name, ID, semester)
        self.course_marks = {}

    def addCourseWithMarks(self, *args):
        course_count = len(args) // 2
        for i in range(course_count):
            course_name = args[i * 2]
            mark = args[i * 2 + 1]
            self.course_marks[course_name] = mark

    def showGPA(self):
        total_credits = 0
        total_marks = 0

        for course, mark in self.course_marks.items():
            credits = 3
            total_credits += credits
            total_marks += self.calculateGradePoint(mark) * credits
            print(f"{course}: {self.calculateGradePoint(mark)}")

        gpa = total_marks / total_credits
        print(f"GPA of {self.name} is: {gpa:.2f}")

    def calculateGradePoint(self, mark):
        if mark >= 85:
            return 4.0
        elif 80 <= mark <= 84:
            return 3.3
        elif 70 <= mark <= 79:
            return 3.0
        elif 65 <= mark <= 69:
            return 2.3
        elif 57 <= mark <= 64:
            return 2.0
        elif 55 <= mark <= 56:
            return 1.3
        elif 50 <= mark <= 54:
            return 1.0
        else:
            return 0.0


# Driver Code
Bob = CSEStudent("Bob", "20301018", "Fall 2020")
Carol = CSEStudent("Carol", "16301814", "Fall 2020")
Anny = CSEStudent("Anny", "18201234", "Fall 2020")

print("#########################")
print(Bob.Details())
print("#########################")
print(Carol.Details())
print("#########################")
print(Anny.Details())
print("#########################")

Bob.addCourseWithMarks("CSE111", 83.5, "CSE230", 73.0, "CSE260", 92.5)
Carol.addCourseWithMarks("CSE470", 62.5, "CSE422", 69.0, "CSE460", 76.5, "CSE461", 87.0)
Anny.addCourseWithMarks("CSE340", 45.5, "CSE321", 95.0, "CSE370", 91.0)

print("----------------------------")
Bob.showGPA()
print("----------------------------")
Carol.showGPA()
print("----------------------------")
Anny.showGPA()
```

```python
class Transport:
    total_traveller = 0

    def __init__(self, name, fare):
        self.name = name
        self.baseFare = fare

    def __str__(self):
        s = "Name: " + self.name + ", Base fare: " + str(self.baseFare)
        return s

    def addPassengerWithBags(self, *args):
        passenger_count = len(args) // 2
        Transport.total_traveller += passenger_count
        print(f"Base-fare of {self.name} is {self.baseFare} Taka")
        print("===============================")
        print("===============================")
        print(f"Name: {self.name}, Base fare: {self.baseFare}")
        print(f"Total Passenger(s): {passenger_count}")

        for i in range(passenger_count):
            name = args[i * 2]
            bag_count = args[i * 2 + 1]

            fare = self.calculateFare(bag_count)
            print(f"Passenger details:")
            print(f"Name: {name}, Fare: {fare}")

        print("===============================")

    def calculateFare(self, bag_count):
        if bag_count <= 2:
            return self.baseFare
        elif 3 <= bag_count <= 5:
            return self.baseFare + 60
        else:
            return self.baseFare + 105


class Bus(Transport):
    def __init__(self, name, fare):
        super().__init__(name, fare)


class Train(Transport):
    def __init__(self, name, fare):
        super().__init__(name, fare)


# Driver Code
t1 = Bus("Volvo", 950)
print("=================================")
t1.addPassengerWithBags("David", 6, "Mike", 1, "Carol", 3)
print("=================================")
print(t1)
print("=================================")

t2 = Train("Silk City", 850)
print("=================================")
t2.addPassengerWithBags("Bob", 2, "Simon", 4)
print("=================================")
print(t2)
print("=================================")
print("Total Passengers in Transport:", Transport.total_traveller)
```

```python
import math

class Vector3D:
    def __init__(self, Xcomponent, Ycomponent, Zcomponent, vec_type='Default'):
        self.Xcomponent = Xcomponent
        self.Ycomponent = Ycomponent
        self.Zcomponent = Zcomponent
        self.vec_type = vec_type

    def __str__(self):
        return f"{self.Xcomponent}i + {self.Ycomponent}j + {self.Zcomponent}k"

    def calculate_length(self):
        return math.sqrt(self.Xcomponent**2 + self.Ycomponent**2 + self.Zcomponent**2)

    def calculate_unit_vector(self):
        length = self.calculate_length()
        if length == 0:
            return "Cannot calculate unit vector for zero vector"
        return Vector3D(
            self.Xcomponent / length,
            self.Ycomponent / length,
            self.Zcomponent / length,
            'unit_vector'
        )

    def __add__(self, other):
        if isinstance(other, Vector3D):
            return Vector3D(
                self.Xcomponent + other.Xcomponent,
                self.Ycomponent + other.Ycomponent,
                self.Zcomponent + other.Zcomponent,
                'resultant_vector'
            )
        else:
            return "Different type of vectors cannot be added"

    def __sub__(self, other):
        if isinstance(other, Vector3D):
            return Vector3D(
                self.Xcomponent - other.Xcomponent,
                self.Ycomponent - other.Ycomponent,
                self.Zcomponent - other.Zcomponent,
                'resultant_vector'
            )
        else:
            return "Different type of vectors cannot be subtracted"

    def __mul__(self, other):
        if isinstance(other, Vector3D):
            return self.Xcomponent * other.Xcomponent + self.Ycomponent * other.Ycomponent + self.Zcomponent * other.Zcomponent
        else:
            return "Different type of vectors cannot be multiplied"

    def is_ortho(self, other):
        if isinstance(other, Vector3D):
            return self * other == 0
        else:
            return "Different type of vectors"

# Driver Code
force1 = Vector3D(1, 3, 5, 'force')
print('---------------------------------------')
print(force1)
print('---------------------------------------')
print('length of force1 vector is: ', force1.calculate_length())
force1_unit_vector = force1.calculate_unit_vector()
print('unit vector of force1 vector is: ', force1_unit_vector)
print('---------------------------------------')
displacement1 = Vector3D(5, -5, 2, 'displacement')
print(displacement1)
print('----------------------------------------')
summ = force1 + displacement1
print('result of addition:', summ)
print('----------------------------------------')
force2 = Vector3D(7, 3, -2, 'force')
summ = force1 + force2
print('result of addition:', summ)
print('----------------------------------------')
diff = force1 - force2
print('result of subtraction:', diff)
print('----------------------------------------')
work = force1 * displacement1
print('work done by force1 and displacement1 is:', work)
print('-----------------------------------------------')
print('force1 and displacement1 are orthogonal:', force1.is_ortho(displacement1))
print('force2 and displacement1 are orthogonal:', force2.is_ortho(displacement1))
```

```python
class Magical_SportsPerson:
    database = {}
    playerNo = 0

    def __init__(self, name, team, jerseyNo):
        self.name = name
        self.team = team
        self.jerseyNo = jerseyNo

    def __str__(self):
        return "Player ID:{},\nName:{},\nTeam:{},\nJersey No:{},\nGoals Scored:{},\nRetirement date:{}".format(
            self.__class__.playerNo + 1, self.name, self.team, self.jerseyNo, 'Not yet retired'
        )

class Quidditch_Player(Magical_SportsPerson):
    def __init__(self, name, team, jerseyNo, goalsScored):
        super().__init__(name, team, jerseyNo)
        self.goalsScored = goalsScored
        Magical_SportsPerson.playerNo += 1
        Magical_SportsPerson.database[f"{Magical_SportsPerson.playerNo}{self.name.split()[0][0]}{self.name.split()[1][0]}{self.jerseyNo}"] = [
            self.name, self.team, self.jerseyNo, self.goalsScored, 'Not yet retired'
        ]

    def retirePlayer(self, retirementDate):
        player_id = f"{Magical_SportsPerson.playerNo}{self.name.split()[0][0]}{self.name.split()[1][0]}{self.jerseyNo}"
        Magical_SportsPerson.database[player_id][-1] = retirementDate

    @classmethod
    def createPlayer(cls, name, team, jerseyNo, goalsScored, retirementDate):
        player = cls(name, team, jerseyNo, goalsScored)
        player.retirePlayer(retirementDate)
        return player

# Driver Code
print("Number of players:", Magical_SportsPerson.playerNo)
print("Player Database:", Magical_SportsPerson.database)
print("#################################")
p1 = Quidditch_Player("Harry Potter", "Gryffindor", 8, 523)
print("------Details of the player------")
print(p1)
print("#################################")
p2 = Quidditch_Player("Ronald Weasley", "Gryffindor", 13, 5)
print("------Details of the player------")
print(p2)
print("#################################")
p3 = Quidditch_Player.createPlayer("George Weasley", "Gryffindor", 12, 11, "11 Magical Year, 1886")
print("------Details of the player------")
print(p3)
print("#################################")
print("Number of players:", Magical_SportsPerson.playerNo)
print("Player Database:", Magical_SportsPerson.database)
```

```python
class User:
    def __init__(self, name, location, ride_type="Single"):
        self.name = name
        self.location = location
        self.ride_type = ride_type
        self.boarded = False

    def status(self):
        if not self.boarded:
            print(f"Status: {self.name} is looking for a {self.ride_type.lower()} ride!")
        else:
            print(f"Status: {self.name} boarded in car {self.boarded.car_number}!")

class Uber:
    def __init__(self, car_number, ride_type, *routes):
        self.car_number = car_number
        self.ride_type = ride_type
        self.routes = routes
        self.passengers = []

    def details(self):
        print(f"Car number: {self.car_number}\nType: {self.ride_type}\nRoutes: {' --> '.join(self.routes)}")

    def pick(self, *users):
        for user in users:
            if not user.boarded:
                if self.ride_type == "Single" or set(user.location) & set(self.routes):
                    print(f"{user.name} has been picked up.")
                    user.boarded = True
                    user.boarded_car = self
                    self.passengers.append(user)
                else:
                    print(f"{user.name} is looking for a different ride.")
            else:
                print(f"{user.name}'s destination is different from this car's route.")


# Driver Code
user1 = User("Brooks", "Banani", "Shared")
user2 = User("Jocelyn", "Uttara")
user3 = User("Robert", "Gulshan", "Shared")
user4 = User("Langdon", "Mohakhali", "Shared")

user1.status()
user2.status()
user3.status()
user4.status()
print("----------------------------------")
car1 = Uber("0K32BH", "Shared", "Mohakhali", "Banani", "Nikunja", "Uttara")
car1.details()
print("----------------------------------")
car1.pick(user1, user2, user3, user4)
print("----------------------------------")
user1.status()
user2.status()
user3.status()
user4.status()
print("----------------------------------")
car2 = Uber("5GD2BD", "Single", "Uttara")
car3 = Uber("4T12FR", "Shared", "Gulshan", "Bashundhara")
car2.details()
car3.details()
print("----------------------------------")
car2.pick(user2, user3)
print("----------------------------------")
car3.pick(user3)
print("----------------------------------")
user2.status()
user3.status()
```
